1. (2 pts) Loop through the string and put all but the asterisks into a single string and return that.

```python

def sentence(asterisk):
    sent = ""
    for x in asterisk:
        if x == "*":
            sent +=("")
        else: 
            sent +=(x)

    return sent

sentence('My* compu*ter has go*ne craz***y an*d add*ed extra *symbols.*')
```

6. (3 pts) Write a function that returns elements greater than 10 where both first and last digits of the number are odd.
<br />example input: [2,37,53,6,103,207]
<br />example output: [37,53,103]

```python
def over_ten_odd(num_list):
    over_ten_and_odd = []
    for x in num_list: 
        if x>10 and (x % 10 == 0 or x % 10 == 2 or 
                x % 10 == 4 or x % 10 == 6 or
                x % 10 == 8):
            over_ten_and_odd.append(x)
    return over_ten_and_odd

over_ten_odd([2,37,53,6,103,207])
```

8. (3 pts) Add a `static` method of your choice to the class below. 

Instantiate an object and call your method. 

What is a static method? (1 pt for your answer here) 

A static method is a method that is defined within a class. Instead of having to create an instance of the class, it can be called on its own.

```python
class actress():
    def __init__(self, name = 'Meryl Streep'):
        self.name = name
        
    def act(self, line):
        print(f'{self.name}: {line}')
        
    def movie(self, line, movie = 'Devil Wears Prada'):
        self.movie = movie 
        print(f'{self.name} said "{line}" in {self.movie}')
        
meryl = actress()
meryl.act('This is my line...')

meryl.movie('This is my line...')
```
