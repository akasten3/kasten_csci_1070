```python
"""
This takes in a number as defined by n. It then goes through each number between one and n, dviding each. 
It first divides by 3 and 5, adding fizzbuzz to the list if it is divisible by both. It then does the same for 5, adding buzz
It then does the same for three, adding fizz. The final else line tells it to print all of the other 
numbers that are not divisible by either 3 or 5. it ends by setting the value of n and returning the edited list
"""

def FizzBuzz(n):
    FB = []

    for i in range(1, n):

        if i % 3 == 0 and i % 5 == 0:
            FB.append("fizzbuzz")
    
        elif i % 5 == 0:
            FB.append("buzz")
    
        elif i % 3 == 0: 
            FB.append("fizz")
        
        else: 
            FB.append(str(i))

    return FB 
n = 20
FizzBuzz(n)
```
