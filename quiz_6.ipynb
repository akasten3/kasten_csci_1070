```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

titanic_df = pd.read_csv("titanic.csv")
```

```python
titanic_df.dropna()
titanic_df.drop_duplicates(inplace = True)
```

```python
sex_map = {'male':0, 'female':1}
titanic_df['Sex'] = titanic_df['Sex'].map(sex_map)
titanic_df
```

```python
X = titanic_df.drop('Survived', axis=1)
y = titanic_df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
print(y_pred)

#i'm not really sure what went wrong in this cell
```

```python
print(knn.score(X_test, y_test))
```

```python
titanic_pred_df = pd.DataFrame(X_test, columns=titanic_df.columns[0:8])

titanic_pred_df['Actual_Outcome'] = y_test.tolist()

titanic_pred_df['Predicted_Outcome'] = y_pred.tolist()

titanic_pred_df.head()
```
