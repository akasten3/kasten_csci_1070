Extra credit (2 pts): Write a function to find the deepest level of nested brackets
<br/>Sample Input: {"a":1,"b":2,"c":{"d":3,"e":4,"f":{"g":5,"h":6}}}
<br/>Sample Output: There are 3 nested layers

```python
def nested_brackets(input):
    a = 0
    for x in (input):
        if x == "{":
            a += 1
        else: 
            a += 0
    return (f'There are {a} nested brackets')

nested_brackets('{"a":1,"b":2,"c":{"d":3,"e":4,"f":{"g":5,"h":6}}}')

```
