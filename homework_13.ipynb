A neural network is a model tmimics the workings of a human brain. It goes through a series of "thoughts" to recognize and analyze patterns. It can be organized as a flow chart. 

the general steps to building a neural network are: 
- defining the structure of the model
- adding a line that tells the children of the model which parents to call
- applying an activation function
- creating an instance of the model
- defining the loss function and optimizer
- minimizing the loss function using either forward or back propagation

you generally check the performance of a neural network with cross entropy. This is because the model is looking for patterns and classifying the data based on those patterns.

```python
from sklearn.preprocessing import LabelEncoder

class_labels = LabelEncoder()
credit_df['Gender'] = class_labels.fit_transform(credit_df['GENDER'].values)
credit_df['Car?'] = class_labels.fit_transform(credit_df['Car_Owner'].values)
credit_df['Property?'] = class_labels.fit_transform(credit_df['Propert_Owner'].values)
credit_df['Income_Type'] = class_labels.fit_transform(credit_df['Type_Income'].values)
credit_df['Education'] = class_labels.fit_transform(credit_df['EDUCATION'].values)
credit_df['Relationship_Status'] = class_labels.fit_transform(credit_df['Marital_status'].values)
credit_df['Housing'] = class_labels.fit_transform(credit_df['Housing_type'].values)
credit_df
```

```python
credit_df = credit_df[['Ind_ID', 'CHILDREN', 'Annual_income', 'Birthday_count', 'Employed_days', 'Mobile_phone',
       'Work_Phone', 'Phone', 'EMAIL_ID', 'Family_Members', 'Gender', 'Car?', 'Property?', 'Income_Type', 
       'Education', 'Relationship_Status', 'Housing']]
credit_df
```

```python
credit_df = credit_df.dropna()
credit_df = credit_df.drop_duplicates()
credit_df
```

```python
#does education impact income type?
credit_df['Income_Type'].to_numpy()

import torch
from sklearn.model_selection import train_test_split
X = credit_df.drop('Income_Type', axis = 1).values
y = credit_df['Income_Type']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, stratify = y)
```

```python
import torch.nn as nn
import torch.nn.functional as F 
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)

y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)

print(X_train)

#not really sure what happened here, but i kept going
```

```python
class ANN_Model(nn.Module):
    def __init__(self, input_features=8, 
                 hidden1=20, hidden2=20, 
                 out_features=2):
        super().__init__()
        self.layer_1_connection = nn.Linear(input_features, hidden1)
        self.layer_2_connection = nn.Linear(hidden1, hidden2)
        self.out = nn.Linear(hidden2, out_features)

    def forward(self, x):
        x = F.relu(self.layer_1_connection(x))
        x = F.relu(self.layer_2_connection(x))
        x = self.out(x)
        return x
```

```python
torch.manual_seed(50)
ann=ANN_Model()
```

```python
loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(ann.parameters(), lr = 0.01)
```

```python
final_loss = []
n_epochs = 500
for epoch in range(n_epochs):
    y_pred = ann.forward(X_train)
    loss = loss_function(y_pred, y_train)
    final_loss.append(loss)

    if epoch % 10 == 1:
        print(f'Epoch number: {epoch} with loss {loss}')

    optimizer.zero_grad() 
    loss.backward()
    optimizer.step()
```
