Deductive reasoning is concluding something based on facts that have already been proved. For instance, I know that all dogs are mammals. My pet is a dog. Therefore, my pet must be a mammal. 

Inductive reasoning is concluding something based off of an observation. For instance, if all of the swans I have ever seen are white, I might recognize that pattern and assume that every swan in the world is white.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = adult_df.drop('income', axis=1)
y = adult_df['income']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, 
                                                    random_state=50, stratify=y)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
adult_df.columns
```

## DECISION TREE

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

features = ['age', 'workclass', 'fnlwgt', 'education', 'eduction_num',
       'marital_status', 'occupation', 'relationship', 'race', 'sex',
       'capital_gain', 'capital_loss', 'hours_per_week', 'native_country']

X = adult_df[features]
y = adult_df['income']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

### Tuning: 
I chose income as my y because I think it connects to all of the rest of the columns the most strongly. I think that each of the other columns would be a large factor in determining how much money someone makes. There were not really any other categories that I thought could have that kind of connection/correlation with the rest of the categories, so I wanted to see that relationship with income. 

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=400, random_state=50)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
rf_tree = rf.estimators_[4]
tree.plot_tree(rf_tree)
plt.savefig("rf_tree.svg")
```
