### 1. (5 pts) In 250-500 words, write a thorough explanation in markdown in your notebook of when to use ML and when not to. Do NOT go over the 500-word limit. It is essential to express this concisely. You may need to do additional research on this than what exists in the readings.

Machine Learning is very helpful when you have a large, unbiased data set. It can help you see trends in the data and help you predict what the future data points will be. It can even help you make assumptions about the whole population. If you have a model that is well-fit to your data, you can predict these things with a great deal of accuracy. However, a too-small data set can cause you to develop an overfitting model that doesn’t predict very well. The model will adapt to the few data points in the set, creating a function that predicts well for those specific points. But, if those data points aren’t representative of the whole data set (i.e. if they are outliers), then the model also won’t be representative of the whole data set. While it may seem well fitting for those points, it might be very far off when predicting future points. A similar problem can occur when using biased data. For instance, if the data is only from a few people who belong to a specific group, it is not going to be very accurate to the whole population. The key is to use a data set that represents the population wholly enough to create a model that accurately predicts the data of the population. If your data set is true to the population, and your model is well-fit to that data set, then machine learning can very accurately predict. But if your data set isn’t very representative, you are likely to end up with a model that is very ill-fitting to the population.
